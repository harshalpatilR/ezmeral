### Function to interact with LLM Model - QA 

```python
# Function to interact with LLM model

def answer_question_with_llama(question: str):
    print(f"Question: {question}")
    BODY_DATA = {
        "model": "meta/llama-3.1-8b-instruct",
        "messages": [
            { "role": "system", "content": "You are a QA assistant respond to user questions " }, 
            { "role": "user", "content": f"{question}"}
        ],
        "max_tokens":1024
    }

    chat_response = requests.post(f"{MODEL_ENDPOINT}{API_PATH}",headers=headers, data=json.dumps(BODY_DATA), verify=False)
    data_from_text = json.loads(chat_response.text)
    out_response=(data_from_text["choices"][0]["message"]["content"])

    #full_response = chat_response.choices[0].message  # This calls the custom Llama-3.1.8B LLM
    
    # Display only the  answer
    #print(f"Answer from Llama-3.18B: {full_response.content}")
    print(f"Answer from LLama-3.18B:{out_response}")
```
