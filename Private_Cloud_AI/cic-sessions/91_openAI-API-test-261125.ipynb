```python
import openai
import os
import requests
import json
```

```python
############# INPUTS ##################
# add /v1 to the API endpoint
#mlis_api_endpoint = "https://qwen-25-72b-it-predictor-harshal-patil-g-308e4e04.ai-application.imid.local/v1"
mlis_api_endpoint = "https://qwen-3-coder-30b-predictor-harshal-patil-g-308e4e04.ai-application.imid.local/v1"

# upload file given by GenAI screen
#token_file="apitoken.txt"
token_file="qwen3-coder-token.txt"

# not required
#mlis_model = "Qwen/Qwen2.5-72B-Instruct-AWQ"
```

```python
# Get these details from MLIS model deployment
try:
    with open(token_file,"r") as f:
        mlis_api_key = f.readline()
        print(f"{token_file}: API Key read from file successfully")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

```python
#### GET MODEL ID - ignores certificate
try:
    headers = {
    'Authorization': f'Bearer {mlis_api_key}',
    'Content-Type': 'application/json' # Often useful to specify content type as well
    }
    
    response = requests.get(f"{mlis_api_endpoint}/models", headers=headers, verify=False)
    
    data = response.json()
    mlis_model=data["data"][0]["id"]
    print(f"Model ID running: {mlis_model}")

    
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

```python
# Ignore certificate checking - for self signed certs
from openai import OpenAI
import httpx
httpx_client = httpx.Client(http2=True, verify=False)

client = OpenAI(
    api_key=mlis_api_key,
    base_url= mlis_api_endpoint,
    http_client=httpx_client
)
```

```python
# stream response with OpenAI API
try:
    stream = client.chat.completions.create(
        model= mlis_model,  
        messages=[
            {"role": "user", "content": "Write b-tree algorithm"}
        ],
        stream=True  # Enable streaming
    )
    for chunk in stream:
        if chunk.choices:
            content = chunk.choices[0].delta.content
            if content is not None:
                print(content, end="", flush=True)
    print()  # Add a newline at the end
except openai.APIError as e:
    print(f"An API error occurred: {e}")

```

```python
# chat completion entire response in one shot
try:
    response = client.chat.completions.create(
        model= mlis_model, # Or a similar Llama 3 model on Together AI
        messages=[
            {"role": "user", "content": "Write python script to call OpenAI API"}
        ],
    )
    print(response.choices[0].message.content)
except openai.APIError as e:
    print(f"An API error occurred: {e}")
```
