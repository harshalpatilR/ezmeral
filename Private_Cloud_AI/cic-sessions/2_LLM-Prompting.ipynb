```python
# Function to interact with LLM model

def promptengg_with_llama(prompt: str):
    print(f"Prompt: {prompt}")
    messages = [
     {"role": "user", "content": f"{prompt}"}
    ]
    
    chat_response = client.chat.completions.create(
         model="meta/llama-3.1-8b-instruct",
         messages=messages,
         max_tokens=500,
         stream=False
    )
    # Pass the question directly to the model (no prompt template used)
    full_response = chat_response.choices[0].message  # This calls the custom Llama-3.1.8B LLM
    
    # Display only the  answer
    print(f"Answer from Llama-3.18B: {full_response.content}")
```

#### System Prompting

```python
#Function for System Prompting 
# Function to interact with LLM model

def promptengg_with_llama_sys(sysprompt: str , prompt: str):
    print(f"System Prompt: {sysprompt}")
    print(f"Prompt: {prompt}")

    messages=[
            {"role": "system", "content": sysprompt},
            {"role": "user", "content": prompt}
        ]   

    chat_response = client.chat.completions.create(
         model="meta/llama-3.1-8b-instruct",
         messages=messages,
         max_tokens=500,
         stream=False
    )
    # Pass the question directly to the model (no prompt template used)
    full_response = chat_response.choices[0].message  # This calls the custom Llama-3.1.8B LLM
    
    # Display only the  answer
    print(f"Answer from Llama-3.18B: {full_response.content}")
```

```python
# Define the System prompt question 
prompt = "How are you ? How is the weather today ?"
```

```python
# Define the System prompt
sys_prompt = "You are a professional language translator. Translate the following text from English to French and only do transalation"
```

```python
# Generate and print the final answer
promptengg_with_llama_sys(sys_prompt,prompt)
```

```python
# Generate and print the final answer , without system prompt 
promptengg_with_llama(prompt)
```

#### Contextual Prompting

```python
context = {
        "tenure": "2 years",
        "account_type": "Postpaid",
        "sim_swap": "Yes (last 24 hours)",
        "device_change": "New device detected",
        "location_change": "Country changed suddenly",
        "time_of_activity": "02:30 AM",
        "past_fraud": "None"
}

transaction = "Request to reset voicemail PIN"
```

```python
prompt = f"""
        You are a telecom fraud detection system.

        RULES:
        - Classify fraud risk accurately
        - Be conservative: if signals are suspicious, mark as High risk
        - Do not explain telecom basics
        - Do not add assumptions beyond the context

        CONTEXT:
        Customer Tenure: {context['tenure']}
        Account Type: {context['account_type']}
        Recent SIM Swap: {context['sim_swap']}
        Device Change: {context['device_change']}
        Location Change: {context['location_change']}
        Time of Activity: {context['time_of_activity']}
        Past Fraud History: {context['past_fraud']}

        TRANSACTION:
        {transaction}

        TASK:
        Classify the fraud risk as one of:
        - Low
        - Medium
        - High

        OUTPUT FORMAT (strict):
        Risk: <Low|Medium|High>
        Reason: <one short sentence>
        """
```

```python
# Generate and print the final answer
promptengg_with_llama(prompt)
```

#### Role Prompting

```python
context = {
        "plan": "Unlimited 5G",
        "location": "Auckland",
        "device": "5G Smartphone",
        "network_status": "Planned maintenance in the area",
        "recent_complaints": "Network issues reported twice last month"
}

customer_message = "My internet is very slow today. This keeps happening."
```

```python
prompt = f"""
        You are a telecom customer support agent working for a mobile network operator.

        ROLE BEHAVIOR:
        - Be polite, empathetic, and professional
        - Use simple, non-technical language
        - Acknowledge the customer issue before explaining
        - Do not speculate or blame
        - Do not promise refunds or compensation

        CUSTOMER CONTEXT:
        Plan: {context['plan']}
        Location: {context['location']}
        Device Type: {context['device']}
        Network Status: {context['network_status']}
        Recent Complaints: {context['recent_complaints']}

        CUSTOMER MESSAGE:
        {customer_message}

        TASK:
        Respond to the customer using the context provided.

        RESPONSE FORMAT:
        - Acknowledgement
        - Explanation
        - Next Step
        """
    
```

```python
# Generate and print the final answer
promptengg_with_llama(prompt)
```

#### Chain of Thought Prompting

```python
context = {
        "technology": "5G NSA",
        "time": "Evening peak hours",
        "area": "Urban residential area",
        "condition": "High network usage and frequent network switching"
}

customer_issue = "My calls keep dropping and the internet is very slow every evening."
```

```python
prompt = f"""
        You are a telecom customer support assistant.

        INSTRUCTIONS:
        - Analyze the situation step by step internally
        - show your internal reasoning in the end of explanation
        - Explain the issue in simple, non-technical language
        - Be calm, reassuring, and concise

        NETWORK CONTEXT:
        Technology: {context['technology']}
        Time of Issue: {context['time']}
        Area Type: {context['area']}
        Network Condition: {context['condition']}

        CUSTOMER ISSUE:
        {customer_issue}

        TASK:
        Provide a clear explanation the customer can understand.

        OUTPUT REQUIREMENTS:
        - 3 to 4 short sentences
        - No technical jargon
        - No bullet points
        - No internal network terms
        - internal reasoning
        """
    
```

```python
# Generate and print the final answer
promptengg_with_llama(prompt)
```
